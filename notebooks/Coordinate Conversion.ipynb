```python
import pandas as pd
import numpy as np
from pyproj import Transformer
transformer = Transformer.from_crs("ESRI:102719", "EPSG:4326")
```

```python
df = pd.read_csv("../data/raleigh_commercial_addresses.csv")
sample = df.sample(n=20, random_state=42)
```

```python
sample['latitude'], sample['longitude'] = transformer.transform(
    sample['y'].values,
    sample['x'].values
)
```

```python
print(sample['latitude'].iloc[0], sample['longitude'].iloc[0])
```

```python
df['latitude'], df['longitude'] = transformer.transform(
    df['y'].values,
    df['x'].values
)
print(df['latitude'].iloc[0], df['longitude'].iloc[0])
```

```python
df_clean = df[['Address', 'City Limits', 'latitude', 'longitude']].copy()
df_clean.to_csv('../data/raleigh_address_coordinates.csv', index=False)
```
