```python
import ssl
import certifi
ssl._create_default_https_context = ssl.create_default_context
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from pyproj import Transformer
geolocator = Nominatim(user_agent='raleigh-delivery-optimizer')
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
df = pd.read_csv("../data/raleigh_address_coordinates.csv")
```

```python
depot = {
    'Address': '4520 Bullock Farm Rd',
    'City Limits': 'Raleigh',
    'latitude': None,
    'longitude': None
}
```

```python
location = geolocator.geocode("4520 Bullock Farm Rd, Raleigh, NC 27610")
depot['latitude'] = location.latitude
depot['longitude'] = location.longitude
```

```python
depot_df = pd.DataFrame([depot])
df = pd.concat([depot_df, df], ignore_index=True)
```

```python
check = df[df['Address'].str.contains("Bullock Farm", case=False)]
print(check)
```
